## ⚗️ 1. Creating and Casting Floats

A float is created any time a number contains a decimal point—even `1.0` is a float, not an integer. You can also produce one through transmutation with `float()`.

````python
# Any decimal point creates a float
a = 7.0
b = 7.5
c = .5    # The leading zero is optional

print(type(a), type(b), type(c))  # All <class 'float'>

# True division always returns a float, even if the result is whole
print(10 / 5)   # 2.0  — float, not 2
print(10 / 2)   # 5.0  — float

# Transmuting other types into floats
from_int    = float(7)       # 7.0
from_str    = float("3.14")  # 3.14
from_str_i  = float("10")    # 10.0 — works even without a decimal point

print(from_int, from_str, from_str_i)
````

> [!TIP]
> **Mixing Whole Stones and Flowing Essence:** When you perform arithmetic with one integer and one float, Python automatically upgrades the result to a float. A single drop of Flowing Essence contaminates the whole cauldron!
> ```python
> 3 + 1.0   # → 4.0  (not 4)
> 10 * 0.5  # → 5.0  (not 5)
> ```

## ⚠️ 2. The Precision Curse: Why `0.1 + 0.2 ≠ 0.3`

Here is one of the most famous curses in all of computing sorcery. Prepare yourself:

````python
result = 0.1 + 0.2
print(result)          # 0.30000000000000004  — not 0.3!
print(result == 0.3)   # False
````

### 🔍 Why Does This Happen?

The machine speaks in **binary**—only zeros and ones. Most decimal fractions, like `0.1`, cannot be written exactly in binary, just as the fraction `⅓` cannot be written exactly in decimal (you get `0.3333...` forever).

The machine stores the *closest possible binary approximation*, and when you add two approximations together, a tiny rounding error appears.

This is not a Python bug—it is a fundamental property of how computers store numbers. Every language that uses 64-bit floats (Java, C, JavaScript) has the same "curse."

### 🛡️ How to Guard Against It

**Never compare floats with `==`**. Instead, check if they are *close enough*:

````python
import math

a = 0.1 + 0.2
b = 0.3

# The unsafe way — can silently fail!
print(a == b)                   # False

# The safe way — check they are close enough
print(math.isclose(a, b))       # True

# For currency or critical precision, use the decimal module
from decimal import Decimal
print(Decimal("0.1") + Decimal("0.2"))  # 0.3  — exact!
````

## 🔮 3. Special Floats: Infinity and the Void

Floating Essence has two extraordinary forms that exist at the edges of the numerical realm: **Infinity** and **NaN** (Not a Number).

````python
# Infinity — a number larger than any other
positive_infinity = float('inf')
negative_infinity = float('-inf')

print(positive_infinity > 10**100)  # True — larger than a googol!
print(negative_infinity < -10**100) # True
print(positive_infinity + 1)        # still inf

# Also accessible through the math module
import math
print(math.isinf(positive_infinity)) # True

# NaN — The Void — appears from undefined operations
void = float('nan')
print(void)            # nan
print(void == void)    # False — the Void is not even equal to itself!
print(math.isnan(void))  # True — the safe way to detect NaN
````

> [!NOTE]
> `math.inf` is a cleaner alias for `float('inf')` and is the preferred form in modern Python.

## 🏛️ 4. The `math` Module: The Academy's Spell Library

The `math` module is a vast library of ready-made mathematical incantations. You unlock it with `import math`.

````python
import math

# Famous constants
print(math.pi)    # 3.141592653589793  — the Circle Constant
print(math.e)     # 2.718281828459045  — Euler's Number
print(math.inf)   # inf

# Rounding siblings
x = 3.7
print(math.floor(x))  # 3   — always round DOWN to the nearest stone
print(math.ceil(x))   # 4   — always round UP to the nearest stone
print(round(x))       # 4   — round to NEAREST (0.5 rounds to even)

# Powers and roots
print(math.sqrt(144))     # 12.0  — square root
print(math.pow(2, 10))    # 1024.0 — same as 2**10, but always returns float
print(math.log(1000, 10)) # 3.0   — log base 10 of 1000

# Trigonometry (useful for spell trajectories!)
angle_radians = math.radians(45)  # Convert 45 degrees to radians
print(round(math.sin(angle_radians), 4))  # 0.7071
````

## 🪞 5. The Enchanted Mirror: f-string Number Formatting

This is where your power as a Python wizard truly shines. The **Enchanted Mirror** (f-string) does far more than simply show a number—it can *polish* and *reshape* it into any presentation you desire.

The secret lies in a **Format Specification** placed after a colon inside the curly braces:

```python
f"{value:format_spec}"
```

### 🔢 Controlling Decimal Places: `f` (fixed-point)

````python
essence = 123.456789

print(f"Raw:          {essence}")       # 123.456789   — no formatting
print(f"4 decimals:   {essence:.4f}")   # 123.4568     — rounded to 4 places
print(f"2 decimals:   {essence:.2f}")   # 123.46       — the most common form
print(f"1 decimal:    {essence:.1f}")   # 123.5
print(f"0 decimals:   {essence:.0f}")   # 123          — rounded to whole number

# Practical use: displaying a gold price
gold_price = 1837.2914
print(f"Gold price: {gold_price:.2f} coins")
````

### 📏 Controlling Width: Padding and Alignment

You can reserve a minimum number of characters for the number. This is essential for building tidy tables and aligned output.

````python
# Format: {value:width.decimalsf}
spells = [("Fireball", 47.5), ("Blizzard", 8.123), ("Lightning", 103.9)]

print("--- Damage Report ---")
for name, damage in spells:
    # 12 chars wide for the number, 2 decimal places, right-aligned (default)
    print(f"  {name:<12} {damage:8.2f} damage")
# Output:
#   Fireball      47.50 damage
#   Blizzard       8.12 damage
#   Lightning    103.90 damage
````

### 💰 Thousands Separator: `,`

Large numbers become much easier to read with a thousands separator.

````python
dragon_hoard = 1234567.89

print(f"Raw:               {dragon_hoard}")        # 1234567.89
print(f"With separator:    {dragon_hoard:,.2f}")   # 1,234,567.89
print(f"Whole only:        {dragon_hoard:,.0f}")   # 1,234,568

# Works on integers too
total_spells = 9876543
print(f"Spells cast: {total_spells:,}")
````

### 📊 Percentage: `%`

The `%` format type multiplies a float by 100 and adds a `%` symbol—perfect for displaying spell success rates and probability magic.

````python
hit_chance     = 0.9273
critical_chance = 0.042
miss_chance    = 1 - hit_chance

print(f"Hit chance:      {hit_chance:.1%}")      # 92.7%
print(f"Critical chance: {critical_chance:.1%}") # 4.2%
print(f"Miss chance:     {miss_chance:.2%}")     # 7.27%

# You can combine comma separator with percentage
tax_rate = 0.2
print(f"Guild tax: {tax_rate:.0%}")  # 20%
````

### 🔬 Scientific Notation: `e`

When numbers are very large or very small, the `e` format presents them in **Scientific Notation**—the language of astronomers and potion chemists.

````python
star_distance  = 9460730472580800.0  # metres in a light-year
potion_particle = 0.000000000137      # size of a molecule

print(f"Light-year:  {star_distance:.2e}")    # 9.46e+15
print(f"Particle:    {potion_particle:.2e}")   # 1.37e-10

# Capital E gives uppercase exponent marker
print(f"Light-year:  {star_distance:.3E}")    # 9.461E+15
````

### ✨ General Format: `g` (Auto-Smart)

The `g` format is the wise, adaptive wizard of formatting. It automatically chooses fixed-point or scientific notation depending on the number's magnitude, and removes trailing zeros.

````python
# g: uses fixed-point for 'normal' numbers, scientific for extremes
# and strips trailing zeros
print(f"{123456.0:.6g}")        # 123456    — no unnecessary .0
print(f"{1234567.0:.6g}")       # 1.23457e+06  — switches to scientific
print(f"{3.14000:.4g}")         # 3.14      — trailing zeros removed
print(f"{0.000123:.3g}")        # 0.000123  — stays fixed
print(f"{0.0000001234:.3g}")    # 1.23e-07  — switches to scientific
````

### 🗺️ The Complete Format Specification Map

The full format specification follows this pattern — every part is optional:

```
{value : [fill][align][sign][#][0][width][grouping][.precision][type]}
```

| Part | Options | Meaning |
| :--- | :--- | :--- |
| `fill` + `align` | `<`, `>`, `^` | Left, right, or centre align |
| `sign` | `+`, `-`, ` ` | Always show sign, only negative, or space for positive |
| `width` | any integer | Minimum total character width |
| `grouping` | `,` or `_` | Thousands separator |
| `.precision` | `.` + integer | Decimal places (for `f`, `e`) or significant figures (for `g`) |
| `type` | `f`, `e`, `E`, `g`, `%`, `d`, `s` | Presentation type |

### 🧪 Combining Multiple Formatting Options

````python
value = 1234.5678

# Width 15, comma separator, 2 decimal places, right-aligned
print(f"{'Amount':>15}: {value:>15,.2f}")

# Always show the + or - sign
gains = 450.75
loss  = -230.20
print(f"Gain: {gains:+.2f}")
print(f"Loss: {loss:+.2f}")

# Zero-padding (useful for codes and serial numbers)
potion_id = 7.0
print(f"Potion ID: {potion_id:06.1f}")  # 0007.0  — padded with zeros
````

## 🔄 6. Rounding Properly: `round()` in Depth

You have seen `round()` before. Let us examine the subtle magic within it.

````python
# round(number, ndigits)
print(round(3.14159, 2))   # 3.14
print(round(3.14159, 4))   # 3.1416
print(round(3.14159, 0))   # 3.0   — returns a float when given a float
print(round(3.14159))      # 3     — returns an int when ndigits omitted

# Negative ndigits rounds to the LEFT of the decimal point
gold = 14823
print(round(gold, -2))     # 14800  — round to the nearest hundred
print(round(gold, -3))     # 15000  — round to the nearest thousand

# Banker's Rounding: Python rounds 0.5 to the nearest EVEN number
print(round(0.5))   # 0  — rounds to 0 (even)
print(round(1.5))   # 2  — rounds to 2 (even)
print(round(2.5))   # 2  — rounds to 2 (even)
print(round(3.5))   # 4  — rounds to 4 (even)
````

> [!NOTE]
> **Banker's Rounding** (round-half-to-even) reduces cumulative rounding errors over many calculations. It differs from the "always round 0.5 up" rule you may have learned in school. If you need classic rounding, use `math.floor(x + 0.5)` instead.

## 🔀 7. `math.floor()` vs `math.ceil()` vs `round()` vs `int()`

All four can produce whole numbers from floats, but they behave very differently. Choosing the wrong one is a classic curse:

````python
import math

values = [2.1, 2.5, 2.9, -2.1, -2.5, -2.9]

print(f"{'Value':>8}  {'int()':>6}  {'floor':>6}  {'ceil':>6}  {'round':>6}")
print("-" * 46)
for v in values:
    print(f"{v:>8.1f}  {int(v):>6}  {math.floor(v):>6}  {math.ceil(v):>6}  {round(v):>6}")
````

| Function | Behaviour | Returns |
| :--- | :--- | :--- |
| `int(x)` | Truncates towards zero — chops the decimal off | `int` |
| `math.floor(x)` | Always rounds **down** (towards −∞) | `int` |
| `math.ceil(x)` | Always rounds **up** (towards +∞) | `int` |
| `round(x)` | Rounds to nearest (ties go to even) | `int` or `float` |

Notice that for negative numbers, `int()` and `math.floor()` give different answers!
