## 🔢 1. The Count-Controlled Loop: The `for` Spell

The most common way to repeat a spell is the `for` loop. We use this when we know exactly how many times we want the magic to happen. 

Consider this simple incantation that makes a magical cat meow three times:


```python
# A simple for loop that repeats 3 times
for i in range(3):
    print("Meow 🐱")
```

### 📜 Breaking Down the Ritual:
- **`for`**: This starts the Echo Charm.
- **`i`**: This is the **Counter Variable**. It’s like a tiny familiar that keeps track of how many times the loop has run.
- **`in range(3)`**: This tells the loop to run 3 times. In Python, `range(3)` creates a sequence: `0, 1, 2`. 
- **`:`**: The "Gateway" symbol. It tells Python that the repeating incantations are about to start!
- **Indentation**: Notice the `print` is pushed to the right? This "Indented Space" tells Python exactly which lines belong inside the loop.

### 🎛️ Normalizing the Counter for Humans (UX)
The machine always starts counting at `0`. But for humans (or "Users"), seeing `0` can be confusing. We can "Normalize" the counter by adding 1 when we display it:


```python
for i in range(3):
    # We add 1 to 'i' so the user sees "Spell 1", "Spell 2", etc.
    print(f"Casting Spell Number {i + 1}!")
```

### 👤 The "Invisible" Counter (`_`)
Sometimes, you just want the loop to run, and you don't actually care about the number. In these cases, wizards use an underscore `_` as the variable name. This tells other wizards: *"I'm using a counter, but I don't need its value."*


```python
for _ in range(5):
    print("Sparkle! ✨")
```

## 📜 2. Iterating Over Magical Collections

Loops aren't just for counting numbers; they can also walk through lists of items or characters in a string. This is called **Iteration**.

### 🏰 Walking Through a List (Hogwarts Students)
A **List** is a magical satchel that holds many items. We can use a `for` loop to pull each item out one by one.


```python
# A list of students (Initiates)
students = ["Harry", "Hermione", "Ron", "Neville", "Luna"]

print("The Great Hall Roll Call:")
for student in students:
    print(f"Welcome, {student}!")
```

### 🧵 Walking Through a String (Letter by Letter)
You can even loop through a "Scroll of Text" (String). The loop will visit every single letter (or "Rune") in the scroll!


```python
spell_word = "ALOHOMORA"

print("Spelling out the Unlocking Charm:")
for letter in spell_word:
    print(f"Rune: {letter}")
```

## 🧪 3. The Condition-Controlled Loop: The `while` Spell

Sometimes, you don't know how many times you need to repeat a spell. You just know you want to keep going **until something changes**. For this, we use the `while` loop.

A `while` loop checks a condition (just like an `if` statement). If the condition is **True**, it runs the code. Then it checks again. And again. It only stops when the condition becomes **False**.


```python
mana_level = 3

print("Charging the Mana Crystal...")
while mana_level > 0:
    print(f"Current Mana: {mana_level}. Casting minor spark...")
    mana_level = mana_level - 1 # We use some mana each time!

print("The crystal is empty. Rest needed.")
```

### 🧭 The `while` Loop Flowchart:
A `while` loop is a circle that only breaks when the condition fails.


```python
import schemdraw
from schemdraw import flow

with schemdraw.Drawing() as d:
    d += flow.Start().label('Start')
    d += flow.Arrow().down(d.unit/2)
    d += (decision := flow.Decision(w=4, h=3, E='YES', S='NO').label('Is Mana > 0?'))
    
    # Yes Path (The Loop)
    d += flow.Arrow().right(d.unit/2).at(decision.E)
    d += (action := flow.Box(w=3, h=1).label('Cast Spark'))
    d += flow.Line().down(d.unit/2).at(action.S)
    d += flow.Line().left(d.unit*2.5)
    d += flow.Arrow().to(decision.N)
    
    # No Path (The Exit)
    d += flow.Arrow().down(d.unit/2).at(decision.S)
    d += flow.Box(w=3, h=1).label('End')
```

## ♾️ 4. The Eternal Flame: Infinite Loops

What if a condition is **always** True? You get an **Infinite Loop**—a spell that never ends! This can be very dangerous (it might freeze your computer tower!), but it is also very useful if you use the right "Emergency Exits."

### 🛑 `break`: The Emergency Stop
The `break` keyword immediately "shatters" the loop and jumps out of it.

### ⏭️ `continue`: The Skip Ritual
The `continue` keyword skips the rest of the current loop and jumps back to the top to start the next turn.


```python
# An infinite loop (Eternal Flame)
print("The Oracle is waiting for the Secret Word...")

while True:
    word = input("Speak the secret word (or type 'exit' to leave): ")
    
    if word.lower() == "exit":
        print("Closing the portal. Goodbye!")
        break # Exit the loop immediately!
        
    if word.lower() == "secret":
        print("You found the hidden chamber! But let's keep the loop going.")
        continue # Skip the 'Unknown word' message and go back to the top
        
    print(f"'{word}' is not the word. Try again!")
```
