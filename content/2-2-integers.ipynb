## 🔬 1. Identifying the Stone: The `type()` Oracle

Before we can work with a magical artifact, a wise wizard identifies what it is. The `type()` oracle reveals the true nature of any value.

````python
whole_stone   = 42
flowing_text  = "42"
flowing_float = 42.0

print(type(whole_stone))   # <class 'int'>
print(type(flowing_text))  # <class 'str'>
print(type(flowing_float)) # <class 'float'>

# They may all look like '42', but they are very different artifacts!
print(whole_stone == flowing_text)  # False — a stone is not a scroll
print(whole_stone == flowing_float) # True  — Python grants this exception
````

## ⚗️ 2. The Art of Type Casting

**Type Casting** is the alchemical process of transforming one type of magical artifact into another. It is one of the most important skills in a wizard's repertoire—and one of the most common sources of curses when done carelessly.

### 📜 Scroll → Stone: `int()` from a String

The `input()` spell always returns a Scroll of Text (string). Before you can do arithmetic with user input, you must transmute it into a Whole Stone.

````python
# The scroll '7' and the stone 7 are not the same!
scroll = "7"
stone  = int(scroll)  # Transmutation!

print(scroll + scroll)  # '77'  — string stitching
print(stone  + stone)   # 14   — arithmetic

# The correct pattern for integer input:
dragon_count = int(input("How many dragons do you see? "))
print(f"Each wizard must fight {dragon_count // 3} dragons.")
````

> [!WARNING]
> **The Impure Scroll Curse:** `int()` will only transmute scrolls that contain a valid whole number. If the scroll contains `"7.5"` or `"seven"` or `" "`, the ritual will fail with a `ValueError`. Always make sure the scroll is pure before casting!

### 💧 Flowing Essence → Stone: `int()` from a Float

You can also transmute a Float into an Integer. But beware—`int()` does not *round* the float. It **truncates** it, slicing away everything after the decimal point like cutting a scroll with a blade. The fractional part is lost forever.

````python
# int() TRUNCATES — it does NOT round
print(int(9.9))   # 9  — not 10! The 0.9 is discarded
print(int(3.1))   # 3
print(int(-2.9))  # -2 — truncates towards zero, NOT towards negative infinity

# Compare with round(), which rounds to the nearest stone
print(round(9.9))  # 10
print(round(9.5))  # 10 (rounds to even)
print(round(9.4))  # 9
````

### 🪨 → 📜 Stone → Scroll: `str()` from an Integer

The reverse ritual — turning a Whole Stone back into a Scroll — uses `str()`. This is needed whenever you want to stitch a number into text using `+` concatenation (though f-strings are almost always easier).

````python
power_level = 9000

# This will raise a TypeError — you cannot stitch a stone to a scroll!
# message = "Power level: " + power_level  # CURSED!

# Correct way with str()
message = "Power level: " + str(power_level)
print(message)  # 'Power level: 9000'

# But really, the Enchanted Mirror is far more elegant:
print(f"Power level: {power_level}")
````

### 🔮 The Full Transmutation Table

| From | To | Ritual | Notes |
| :--- | :--- | :--- | :--- |
| `str` | `int` | `int("42")` | Fails if string is not a pure whole number |
| `float` | `int` | `int(3.9)` | Truncates — **not** rounds |
| `int` | `float` | `float(7)` | Becomes `7.0` |
| `int` | `str` | `str(42)` | Becomes `"42"` |
| `bool` | `int` | `int(True)` | `True` → `1`, `False` → `0` |
| `int` | `bool` | `bool(0)` | `0` → `False`, anything else → `True` |

## ⚔️ 3. Mathematical Incantations: The Operators in Depth

You met the seven operators in the Numerology lesson. Now let us study them with wizard-level precision.

### 🔢 The Operators

````python
a = 17
b = 5

print(f"{a} + {b}  = {a + b}")    # 22  — Addition
print(f"{a} - {b}  = {a - b}")    # 12  — Subtraction
print(f"{a} * {b}  = {a * b}")    # 85  — Multiplication
print(f"{a} / {b}  = {a / b}")    # 3.4 — True Division (ALWAYS returns a float)
print(f"{a} // {b} = {a // b}")   # 3   — Floor Division (drops remainder)
print(f"{a} % {b}  = {a % b}")    # 2   — Modulo (the remainder)
print(f"{a} ** {b} = {a ** b}")   # 1419857 — Exponentiation (17 to the power of 5)
````

> [!TIP]
> **The Modulo Mystery:** The `%` operator is far more powerful than it first appears. It is used everywhere in magic:
> - Check if a number is even: `n % 2 == 0`
> - Wrap numbers around a circle: `angle % 360`
> - Distribute items fairly and find what's left over

### ⚖️ The Order of Operations: The Arcane Hierarchy

When multiple operators appear in one incantation, Python follows the **Arcane Hierarchy** (you may know it as BODMAS or PEMDAS). Operators higher in the hierarchy are resolved first.

| Priority | Operator | Name |
| :---: | :---: | :--- |
| 1 (highest) | `()` | Parentheses — the Binding Circle |
| 2 | `**` | Exponentiation — the Power Rune |
| 3 | `*`, `/`, `//`, `%` | Multiplication and Division family |
| 4 (lowest) | `+`, `-` | Addition and Subtraction |

````python
# Without parentheses — the hierarchy decides
result_a = 2 + 3 * 4   # 3*4 resolved first → 2 + 12 = 14
print(result_a)         # 14

# With parentheses — we override the hierarchy
result_b = (2 + 3) * 4  # (2+3) resolved first → 5 * 4 = 20
print(result_b)          # 20

# A more complex incantation
fireball_damage = 2 ** 3 + 10 // 3 - 1   # → 8 + 3 - 1 = 10
print(f"Fireball damage: {fireball_damage}")  # 10

# When in doubt, use parentheses — it makes the spell easier to read!
fireball_damage_clear = (2 ** 3) + (10 // 3) - 1
print(f"Same result: {fireball_damage_clear}")
````

### 🛠️ Useful Built-in Spells for Integers

Python's Standard Grimoire includes several ready-made spells that work on integers:

````python
# abs() — The Absolute Value Ward (strips the negative sign)
debt = -350
print(abs(debt))          # 350  — now we know the magnitude

# pow(base, exp) — Same as ** but can take a third argument for modulo
print(pow(2, 10))          # 1024
print(pow(2, 10, 1000))    # 24   — (2**10) % 1000, useful in cryptography!

# divmod() — Returns BOTH the quotient AND the remainder as a pair
crystals = 17
wizards  = 5
quotient, remainder = divmod(crystals, wizards)
print(f"Each wizard gets {quotient} crystals, {remainder} left for the master.")

# min() and max() — Find the smallest or largest stone
print(min(3, 7, 1, 9, 4))  # 1
print(max(3, 7, 1, 9, 4))  # 9
````

## 🚫 4. The Forbidden Rune: Why `++` Does Not Exist in Python

Many initiates arrive at the Academy having studied other scrolls—C, Java, or JavaScript—where a spell called `++` exists. This operator adds 1 to a number instantly:

```java
// In Java or C — this works:
int counter = 5;
counter++;  // counter is now 6
```

When they try this in Python, something mysterious happens—**it appears to work, but it doesn't do what you think**:

````python
counter = 5

# This LOOKS like it should increment counter, but...
++counter

print(counter)  # Still 5! Nothing happened.
````

### 🔍 Why Does `++counter` Seem to Work (But Doesn't)?

Python has no `++` operator. What Python actually sees when you write `++counter` is two **unary plus operators** applied in sequence:

```
+(+counter)
```

A unary `+` does nothing—it returns the same number unchanged. So `++counter` simply evaluates to `counter`. It's like writing `+5` instead of `5`. And crucially, **the result is not stored back into `counter`**—it's just discarded.

The same trap exists for `--counter`: it becomes `-(−counter)`, which is just `counter` again.

The Python Arch-Masters made a deliberate choice to leave `++` out of the language. Python has a cleaner, more explicit way to do this—the **Augmented Assignment Runes**.

## ✨ 5. The Augmented Assignment Runes: `+=` and its Kin

Augmented assignment operators are a shorthand for updating a variable's own value. Instead of writing:

```python
mana = mana + 10
```

You can write:

```python
mana += 10
```

Both do exactly the same thing—they read the current value, apply the operation, and **write the result back into the same variable**. The augmented form is simply shorter and more elegant.

Every arithmetic operator has an augmented form:

````python
mana = 100
print(f"Starting mana: {mana}")

mana += 20    # mana = mana + 20  →  Absorb mana from a crystal
print(f"After += 20:  {mana}")   # 120

mana -= 35    # mana = mana - 35  →  Cast a costly spell
print(f"After -= 35:  {mana}")   # 85

mana *= 2     # mana = mana * 2   →  Mana amplification ritual
print(f"After *= 2:   {mana}")   # 170

mana //= 3    # mana = mana // 3  →  Divide between 3 wizards (floor)
print(f"After //= 3:  {mana}")   # 56

mana %= 10    # mana = mana % 10  →  Keep only the remainder
print(f"After %= 10:  {mana}")   # 6

mana **= 2    # mana = mana ** 2  →  Square the remaining power!
print(f"After **= 2:  {mana}")   # 36
````

### 🔄 The Correct Way to Increment and Decrement

Since `++` doesn't exist, Python wizards use `+= 1` and `-= 1` as the standard counting runes. This is the universal convention across all Python scrolls:

````python
# Counting spells cast in a practice session
spells_cast = 0

spells_cast += 1   # Cast first spell
spells_cast += 1   # Cast second spell
spells_cast += 1   # Cast third spell
print(f"Spells cast: {spells_cast}")  # 3

# Counting down a magical timer
countdown = 5

countdown -= 1
countdown -= 1
countdown -= 1
print(f"Countdown: {countdown}")  # 2

# Doubling power each time
power = 1

power *= 2   # 2
power *= 2   # 4
power *= 2   # 8
print(f"Power level: {power}")  # 8
````

### 📋 The Augmented Runes at a Glance

| Rune | Equivalent Long Form | Effect |
| :---: | :--- | :--- |
| `x += n` | `x = x + n` | Add `n` to `x` |
| `x -= n` | `x = x - n` | Subtract `n` from `x` |
| `x *= n` | `x = x * n` | Multiply `x` by `n` |
| `x /= n` | `x = x / n` | Divide `x` by `n` (result is a float) |
| `x //= n` | `x = x // n` | Floor-divide `x` by `n` |
| `x %= n` | `x = x % n` | Replace `x` with the remainder of `x ÷ n` |
| `x **= n` | `x = x ** n` | Raise `x` to the power of `n` |

> [!TIP]
> **`+=` works on strings too!** `message += " more text"` appends to a string. And it works on lists. The augmented runes are more versatile than they look.

## 🧩 6. Putting It All Together: A Complete Example

Let us watch all of today's lessons work together in a single practical spell—a mana management tracker for a wizard duel:

````python
# === THE DUEL MANA TRACKER ===

# Type casting: accept user input as integers
starting_mana  = int(input("Enter starting mana (e.g. 100): "))
spell_cost     = int(input("Spell cost per round (e.g. 13): "))
mana_per_round = int(input("Mana recovered per round (e.g. 5): "))

current_mana = starting_mana

# Simulate 3 rounds using augmented assignment
current_mana -= spell_cost     # Cast spell
current_mana += mana_per_round # Recover
print(f"After round 1: {current_mana} mana")

current_mana -= spell_cost
current_mana += mana_per_round
print(f"After round 2: {current_mana} mana")

current_mana -= spell_cost
current_mana += mana_per_round
print(f"After round 3: {current_mana} mana")

# How many full spells can you still cast?
spells_remaining, leftover = divmod(current_mana, spell_cost)
print(f"\nRemaining full spells: {spells_remaining}")
print(f"Leftover mana: {leftover}")
print(f"Mana as % of starting: {round(current_mana / starting_mana * 100)}%")
````
