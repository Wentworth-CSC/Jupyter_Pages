## Variables
A variable is just a container for a value within your own program.

In your program, you can introduce your own variable in your program or assigning them usng the `input` incantation we've just learned. Considet the following spell:



````python
my_name = "The Python Oracle"
your_name = input("What's your name? ")
print(f"hello, {your_name}, Welcome to pyWhisper - sincerely, {my_name}")
````

Notice that this equal = sign in the middle of `your_name = input("What's your name? ")`?

The equals sign has a special role in programming. It says "I want you to create a brand new box to hold some information I'll want later. Give it the name on the left hand side, and put the result of anything I say on the right hand side in it.  Therefore, the value returned by `input("What's your name? ")` (which will be your name, as you typed it in) is put into a box called `your_name`. As you can see, I (The Python Oracle) put my name into another box called `my_name`.

## A note about variable names
As you write more and more spells, you will want to be careful about the names you choose for your variables. Variable names should be descriptive and meaningful, but also short enough to be easily typed. Avoid using names that are too generic or too long, as this can make your code harder to read and understand. Additionally, be careful not to use names that are already reserved for Python keywords or built-in functions, as this can cause errors in your code.
