## 🛡️ 1. The Simple Guard: The `if` Statement

The most basic decision is the `if` statement. It's like a guard golem standing at a door. It only lets the magic pass if a certain condition is **True**.

### 📜 The Guard's Logic:
If the condition is met, the "indented" incantations inside the block are cast. If not, the golem ignores them completely and the spell continues below.


```python
# The Guard Golem Spell
wand_ready = True

if wand_ready:
    print("The Golem bows and opens the gate.")
    print("You may enter the Sanctum.")

print("The journey continues...")
```

### 🧭 Visualizing the Path:
We can use a **Flowchart** to see how the magic flows through the guard.


```python
import schemdraw
from schemdraw import flow

with schemdraw.Drawing() as d:
    d += flow.Start().label('Start')
    d += flow.Arrow().down(d.unit/2)
    d += (decision := flow.Decision(w=4, h=3, E='YES', S='NO').label('Is Wand Ready?'))
    d += flow.Arrow().down(d.unit/2).at(decision.S)
    d += flow.Box(w=4, h=1).label('Continue Journey')
    d += flow.Arrow().right(d.unit/2).at(decision.E)
    d += (open_gate := flow.Box(w=4, h=1).label('Open Gate'))
    d += flow.Line().down(d.unit*1.1).at(open_gate.S)
    d += flow.Line().left(d.unit*2.5)
    d += flow.Arrow().to(decision.S)
```

## ⚖️ 2. The Fork in the Road: `if-else`

What if you want the golem to do something else if the condition is **False**? For that, we use the `else` ritual. It's an "either-or" choice.


```python
# The House Entrance Spell
is_student = False

if is_student:
    print("Welcome back, Initiate!")
else:
    print("Halt! Only students may enter the Common Room.")
    print("Summoning the Headmaster...")
```

### 🧭 The if-else Flowchart:


```python
with schemdraw.Drawing() as d:
    d += flow.Start().label('Start')
    d += flow.Arrow().down(d.unit/2)
    d += (dec := flow.Decision(w=4, h=3, E='YES', W='NO').label('Is Student?'))
    
    # Yes Path
    d += flow.Arrow().right(d.unit/2).at(dec.E)
    d += (yes_box := flow.Box(w=3, h=1).label('Welcome'))
    
    # No Path
    d += flow.Arrow().left(d.unit/2).at(dec.W)
    d += (no_box := flow.Box(w=3, h=1).label('Halt'))
    
    # Converge
    d += flow.Line().down(d.unit).at(yes_box.S)
    d += (btm := flow.Line().left(d.unit*3.5))
    d += flow.Line().down(d.unit).at(no_box.S)
    d += flow.Line().right(d.unit*3.5)
    d += flow.Arrow().down(d.unit/2).at(btm.end)
    d += flow.Box(w=3, h=1).label('End')
```

## 📜 3. The Examiner's Scroll: `if-elif-else`

Sometimes there are more than two paths. Imagine the **Grand Examiner** grading your magical trials. In the Cambridge International traditions, we have many grades!

We use `elif` (short for "else if") to check multiple conditions one by one.

### 🏫 The Grading Ritual:
- **A*** >= 90
- **A** >= 80
- **B** >= 70
- **C** >= 60
- **D** >= 50
- **E** >= 40
- **U** < 40 (The Unfortunate grade)


```python
# The Examiner's Grading Spell
score = int(input("Enter your trial score (0-100): "))

if score >= 90:
    grade = "A*"
elif score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
elif score >= 40:
    grade = "E"
else:
    grade = "U"

print(f"The Examiner marks your scroll with: {grade}")
```

## 🤝 4. Logical Rituals: `and`, `or`, and the "Harry Trap"

Wizards often need to combine conditions. We use **Logical Operators** for this:
- **`and`**: BOTH conditions must be True.
- **`or`**: At least ONE condition must be True.
- **`not`**: Flips the result (True becomes False).

### ⚠️ The Apprentice's Pitfall
Many initiates try to check if a name is 'Harry' or 'Hermione' like this:
`if name == 'Harry' or 'Hermione':`

**This is a curse!** In Python's eyes, this means:
1. Is `name == 'Harry'`?
2. **OR** is the string `'Hermione'` true? (In Python, non-empty strings are *always* True!)

So this spell will *always* be True, even if your name is Voldemort! 😱

### ✅ The Correct Incantation:
You must be explicit with your comparisons:
`if name == 'Harry' or name == 'Hermione':`


```python
# The House Sorting Spell
student_name = input("Enter student name: ")

if student_name == "Harry" or student_name == "Hermione" or student_name == "Ron":
    house = "Gryffindor"
elif student_name == "Draco":
    house = "Slytherin"
elif student_name == "Luna":
    house = "Ravenclaw"
else:
    house = "Unknown Tower"

print(f"{student_name} is sorted into: {house}")
```

## 🪞 5. The Grand Sorting Portal: The `match` Statement

In recent years (Python 3.10+), a more powerful ritual called **Structural Pattern Matching** was discovered. It is much cleaner than a long list of `elif` statements.

We use the `match` and `case` keywords. We can even use the **Pipe symbol (`|`)** to mean "or" inside a single case!


```python
# The Modern Sorting Spell
name = input("Who approaches the Sorting Hat? ")

match name:
    case "Harry" | "Hermione" | "Ron" | "Neville":
        print("House: Gryffindor! ❤️🦁")
    case "Draco" | "Pansy" | "Crabbe":
        print("House: Slytherin! 💚🐍")
    case "Luna" | "Cho":
        print("House: Ravenclaw! 💙🦅")
    case "Cedric":
        print("House: Hufflepuff! 💛🦡")
    case _: # The 'Wildcard' (like 'else')
        print("The Hat is confused... try a different name!")
```

### 🧭 The Match Flowchart:
A `match` statement is like a multi-way switch.


```python
with schemdraw.Drawing() as d:
    d += flow.Start().label('Start')
    d += flow.Arrow().down(d.unit/2)
    d += (dec := flow.Decision(w=4, h=3).label('Match Name?'))
    
    # Paths
    d += flow.Arrow().right(d.unit/2).at(dec.E).label('Harry|Ron...')
    d += flow.Box(w=3, h=1).label('Gryffindor')
    
    d += flow.Arrow().down(d.unit/2).at(dec.S).label('Luna|Cho')
    d += flow.Box(w=3, h=1).label('Ravenclaw')
    
    d += flow.Arrow().left(d.unit/2).at(dec.W).label('Draco|Pansy')
    d += flow.Box(w=3, h=1).label('Slytherin')
```
